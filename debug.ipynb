```python
import matplotlib.pyplot as plt
import pickle
with open('results/sctm_v1.pkl', 'rb') as f:
    cached_loss_v1 = pickle.load(f)
fig, axs = plt.subplots(3, 4, figsize=(9, 6))
i = 0
for name in cached_loss_v1.keys():
    axs[i // 4, i % 4].plot(cached_loss_v1[name])
    axs[i // 4, i % 4].set_title(name)
    i += 1
plt.tight_layout()
plt.show()
```
